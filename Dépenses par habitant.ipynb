```kotlin
import java.io.File

val data = File("Dépenses par habitant.csv").readLines()
%use kandy
```

```kotlin
val formated = mapOf("Pays" to data.map { it.substringBefore(";") }, "Dépenses" to data.map { it.substringAfter(";").toInt() })
val country = "Pays"<String>()
val spending = "Dépenses"<Int>()

plot(formated) {
    bars {
        x(spending) {
            scale = continuous(0..50000)
            axis.breaks(format = " ")
        }
        y(country)
    }
    layout {
        title = "Dépense de l'État par habitant"
        size = 800 to 600
    }
}
```
